```python
#trips = pd.read_table('rt_trips_DB_2018.txt',sep=';')
trips = pd.read_csv('trips.csv',sep=',')
```

```python
f = open('leavetime_arr.csv')
reader = pd.read_csv(f, sep=',',iterator=True, index_col=0)
loop = True
chunkSize = 500000
leavetime_arr = []
while loop:
    try:
        chunk = reader.get_chunk(chunkSize)
        leavetime_arr.append(chunk)
    except StopIteration:
        loop = False
        print("Iteration is stopped.")
```
